**Generating Historical Correlation Matrices and Heatmap Pots**

```python
exh = 'NSE'
nn = 3
nnn = 3
corr_matrix = pd.DataFrame()
for n in range(nnn,len(index_data)):

    if n == nnn:
        dat = str(index_data[-n:].index[-1]).split(' ')[0]
        correlation_matrix = index_data[-n:].corr(numeric_only = False)
    else:
        dat = str(index_data[-n:-(n-nn)].index[-1]).split(' ')[0]
        correlation_matrix = index_data[-n:-(n-nn)].corr(numeric_only = False)
    lower_triangle = np.tril(correlation_matrix, k=-1)
    li = []
    for i in range(1,lower_triangle.shape[0]):
        for j in range(0,i):
            li.append(lower_triangle[i][j])
    li = pd.DataFrame(li)
    corr_matrix = pd.concat([corr_matrix, pd.DataFrame([dat,(li.sum()/len(li)).values[0] ]).T.rename(columns = {
        0:'date',
        1:'corr_avg'
    })])
    ## corr_matrix consist correlation Average of the lower half of the correlation matrix
    ## uncomment the below code to plot every correlation plots
    # plt.close()
    # if not os.path.exists('plots/'+exh + '_' + str(nn) ):
    #     os.makedirs('plots/'+exh+'_'+str(nn))
    # plt.figure(figsize=(10, 8))
    # plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
    # plt.colorbar(label='Correlation')


    # plt.title('Correlation Matrix of '+ exh +' Indices for, including ' + dat)
    # plt.xticks(ticks=range(len(correlation_matrix)), labels=correlation_matrix.columns, rotation=45)
    # plt.yticks(ticks=range(len(correlation_matrix)), labels=correlation_matrix.index)
    # plt.savefig('plots/'+exh+'_'+str(nn)+'/'+  str(index_data[-n:].index[-1]).split(' ')[0] +'.png')
    # plt.savefig('plots/'+exh+'_'+str(nn)+'/'+  dat +'.png')
    # plt.show()
    print(dat)
    clear_output(wait=True)
clear_output(wait=False)
```

```python
index_data['date'] = [str(i).split(' ')[0] for i in list(index_data.index)]
index_data = index_data.reset_index(drop = True)
```

```python

li_dif1_c = pd.DataFrame()
li_dif2_c = pd.DataFrame()
for i in chaos_corr_li:
    print(i) # 1 ~ up and 0 ~ down
    dif1 = [ 1 if i > 0 else -1
    for i in list((
        index_data.iloc[index_data[index_data['date'] == i].index[0]].drop('date')
        - index_data.iloc[index_data[index_data['date'] == i].index[0] - 1].drop('date')
    )/index_data.iloc[index_data[index_data['date'] == i].index[0] - 1].drop('date'))]
    dif2 = [ 1 if i > 0 else -1
    for i in list((
        index_data.iloc[index_data[index_data['date'] == i].index[0]-1].drop('date')
        - index_data.iloc[index_data[index_data['date'] == i].index[0] - 2].drop('date')
    )/index_data.iloc[index_data[index_data['date'] == i].index[0] - 2].drop('date'))]
    li_dif1_c[i] = dif1
    li_dif2_c[i] = dif2

    clear_output(wait=True)
li_dif1_c.sum().hist(color = 'black')
plt.title("Histogram for Chaos Plots")
plt.xlabel('Sum of Bullish or Bearish Behavior')
plt.ylabel('Number of Days they were Bullish / Bearish')
plt.show()

li_dif2_c.sum().hist(color = 'black')
plt.title("Histogram for Chaos Plots")
plt.xlabel('Sum of Bullish or Bearish Behavior')
plt.ylabel('Number of Days they were Bullish / Bearish')
plt.show()
```

```python
li_dif1_p = pd.DataFrame()
li_dif2_p = pd.DataFrame()
for i in peace_corr_li:
    print(i) # 1 ~ up and 0 ~ down
    dif1 = [ 1 if i > 0 else -1
    for i in list((
        index_data.iloc[index_data[index_data['date'] == i].index[0]].drop('date')
        - index_data.iloc[index_data[index_data['date'] == i].index[0] - 1].drop('date')
    )/index_data.iloc[index_data[index_data['date'] == i].index[0] - 1].drop('date'))]
    dif2 = [ 1 if i > 0 else -1
    for i in list((
        index_data.iloc[index_data[index_data['date'] == i].index[0]-1].drop('date')
        - index_data.iloc[index_data[index_data['date'] == i].index[0] - 2].drop('date')
    )/index_data.iloc[index_data[index_data['date'] == i].index[0] - 2].drop('date'))]
    li_dif1_p[i] = dif1
    li_dif2_p[i] = dif2

    clear_output(wait=True)
li_dif1_p.sum().hist(color = 'black')
plt.title("Histogram for Peace Plots")
plt.xlabel('Sum of Bullish or Bearish Behavior')
plt.ylabel('Number of Days they were Bullish / Bearish')
plt.show()

li_dif2_p.sum().hist(color = 'black')
plt.title("Histogram for Peace Plots")
plt.xlabel('Sum of Bullish or Bearish Behavior')
plt.ylabel('Number of Days they were Bullish / Bearish')
plt.show()
```

```python
import pandas as pd
from datetime import datetime
peace =  pd.DataFrame(li_dif1_p.sum()).rename(columns = {0:'sum_sentiments'})
peace = peace.reset_index().rename(columns = {'index':'date'})
peace = peace[::-1].reset_index(drop=True) # reverse rows for date order
change = 0
no_change = 0
cd = {}
ncd = {}
for i in range(0,len(peace.date)-1):

    # case handling when the next plot is chaos when the current plot is peace.
    dat = peace.iloc[i].date
    chk = datetime.strptime(peace.iloc[i].date,'%Y-%m-%d') + timedelta(days = 1)
    cont = 0
    while True:
      if chk.strftime('%Y-%m-%d') in peace_corr_li:
        break
      if chk.strftime('%Y-%m-%d') in chaos_corr_li:
        cont = 1
        break
      if chk > datetime.strptime(peace.iloc[-1].date,'%Y-%m-%d'):
        break
      chk += timedelta(days = 1)
    if cont == 1:
      continue
    # case handling end
    day = datetime.strptime(peace.iloc[i].date,"%Y-%m-%d").weekday()
    if  day not in cd:
        cd[day] = 0
    if day not in ncd:
        ncd[day] = 0
    if peace.iloc[i].sum_sentiments <= 0:
        if peace.iloc[i + 1].sum_sentiments <= 0 :
            no_change += 1
            ncd[day] += 1
        else:
            cd [day] += 1
            change += 1
    else:
        if peace.iloc[i + 1].sum_sentiments > 0 :
            ncd[day] += 1
            no_change += 1
        else:
            cd[day] += 1
            change += 1
    print(dat)
    clear_output(wait=True)
clear_output(wait=False)
cd = {key: cd[key] for key in sorted(cd.keys())}
ncd = {key: ncd[key] for key in sorted(ncd.keys())}
print('done')
```

# Get CURRENT correlation heatmap matrix of the NSE, BSE market

```python
from datetime import datetime, timedelta
import pandas as pd
import yfinance as yf
from IPython.display import clear_output
import matplotlib.pyplot as plt
import time
```

```python
def create_corr(exh, dtt, tic):
    end_date = dtt + timedelta(days = 1)
    start_date = end_date - timedelta(days = 5)
    # tic = [ i.replace('(','').replace(')','').replace('\xa0','') for i in list(
    #     dff[dff.exchange == exh]['Yfinance']
    # )]
    data = yf.download(tic, interval = '1m', start = start_date, end=end_date)
    yd = data['Adj Close']
    yd = yd.dropna()
    yd['Datetime'] = [str(i) for i in list(yd.index)]
    date_li = list(pd.DataFrame([ str(i).split(' ')[0] for i in list(yd.index)])[0].unique())
    date_li = date_li[-3:]   # taking only 3 previous dyas including today after the market is closed.
    tmp = pd.DataFrame()
    for i in date_li:
        ydd = yd[yd.Datetime.str.contains(i)]
        tmp = pd.concat([tmp,pd.DataFrame(ydd.iloc[-1]).T])
    tmp.drop('Datetime', axis = 1, inplace = True)
    correlation_matrix = tmp.corr(numeric_only= False)
    plt.close()
    plt.figure(figsize=(10, 8))
    plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
    plt.colorbar(label='Correlation')
    plt.title('Correlation Matrix of '+exh+' Indices ' + str(tmp.index[-1]).split(' ')[0] )
    plt.xticks(ticks=range(len(correlation_matrix)), labels=correlation_matrix.columns, rotation=45)
    plt.yticks(ticks=range(len(correlation_matrix)), labels=correlation_matrix.index)
    plt.savefig( exh + '_+ str(tmp.index[-1]).split(' ')[0].replace('-','_') + '.png')
    plt.show()
```
